```python
import pandas as pd
```

```python
df = pd.read_csv(r'/content/sample_data/data1.csv')
```

```python
df.head()
```

```python
df1 = df.iloc[:,0:2]
```

```python
df1.head()
```

```python
from googletrans import Translator, constants
```

```python
translator = Translator()
```

```python
a = []
for i in df1['FULL_TEXT']:
    translation = translator.translate(i, dest="en")
    a.append(translation.text)
```

```python
a
```

```python
len(a)
```

```python
for i in a :
  df1['translated_phrase']= a
```
